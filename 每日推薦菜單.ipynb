## 複習(資料隨機產生、計算)並延伸成多筆資料

```python
#先複習怎麼隨機抽取並儲存答案
import random

friends = {
    "壺阿莉": {"班級": "順天書屋", "興趣": "Python", "年紀": 18},
    "周小餅": {"班級": "順天書屋", "興趣": "Coding", "年紀": 12},
    "林小銘": {"班級": "順天書屋", "興趣": "打籃球", "年紀": 13},
    "林小庭": {"班級": "順天書屋", "興趣": "看書", "年紀": 12},
    "戴阿哲": {"班級": "順天書屋", "興趣": "畫畫", "年紀": 13},
    "張小國": {"班級": "順天書屋", "興趣": "音樂", "年紀": 11},
    "邱阿恩": {"班級": "順天書屋", "興趣": "寫程式", "年紀": 12}
}


def pick_random_friends():
    # 隨機抽三個朋友
    selected = random.sample(list(friends.items()), 3)

    # 用迴圈計算總年紀（直觀版）
    total_age = 0
    for name, info in selected:
        total_age += info["年紀"]  # 每次把朋友的年紀加到 total_age

    # 回傳選到的朋友和總年紀
    return selected, total_age

# 測試
chosen, total = pick_random_friends()
print("抽到的朋友：")
for name, info in chosen:
    print(name, "年紀:", info["年紀"])
print("總年紀:", total)
```

複習完隨機抽取，我們做更進一步的延伸，我們隨機抽取3個人，並重複10次，當作一個值日生表，同時如果我們好奇值日生的平均年紀，可以

```python
# 抽取 10 天資料
records = []

for day in range(1, 11):
    chosen, total = pick_random_friends()
    records.append([day, chosen, total, total / 3])  # 紀錄天數、朋友、總年紀、平均年紀

# 印出結果
for day, chosen, total, avg in records:
    print(f" {day} Day：")
    for name, info in chosen:
        print(f"  {name} 年紀: {info['年紀']}")
    print(f"  總年紀: {total}，平均年紀: {avg:.1f}")
    print("-" * 30)
```

##資料整理

```python
import pandas as pd

# 轉換 records 成 DataFrame
data = []

for day, chosen, total, avg in records:
    # 將三位朋友的名字分欄
    names = [name for name, info in chosen]
    data.append([day, names[0], names[1], names[2], total, avg])

# 定義欄位名稱
columns = ["日期", "朋友1", "朋友2", "朋友3", "總年紀", "平均年紀"]

df = pd.DataFrame(data, columns=columns)
df
```

## 建個函式幫我們隨機創出三餐的menu，並記錄卡路里

```python

def generate_daily_meal():
    早餐 = {
        "五穀雜糧": random.choice(list(food_calories["五穀雜糧"].keys())),
        "蛋白質": random.choice(list(food_calories["蛋白質"].keys())),
        "乳製品": random.choice(list(food_calories["乳製品"].keys())),
        "水果": random.choice(list(food_calories["水果"].keys()))
    }

    午餐 = {
        "五穀雜糧": random.choice(list(food_calories["五穀雜糧"].keys())),
        "蛋白質": random.choice(list(food_calories["蛋白質"].keys())),
        "蔬菜": random.choice(list(food_calories["蔬菜"].keys())),
        "水果": random.choice(list(food_calories["水果"].keys())),
        "油脂": random.choice(list(food_calories["油脂"].keys()))
    }

    晚餐 = {
        "五穀雜糧": random.choice(list(food_calories["五穀雜糧"].keys())),
        "蛋白質": random.choice(list(food_calories["蛋白質"].keys())),
        "蔬菜": random.choice(list(food_calories["蔬菜"].keys())),
        "水果": random.choice(list(food_calories["水果"].keys())),
        "油脂": random.choice(list(food_calories["油脂"].keys()))
    }

# 計算每餐的總熱量
    早餐熱量 = (food_calories["五穀雜糧"][早餐["五穀雜糧"]] +
                food_calories["蛋白質"][早餐["蛋白質"]] +
                food_calories["乳製品"][早餐["乳製品"]] +
                food_calories["水果"][早餐["水果"]])

    午餐熱量 = (food_calories["五穀雜糧"][午餐["五穀雜糧"]] +
                food_calories["蛋白質"][午餐["蛋白質"]] +
                food_calories["蔬菜"][午餐["蔬菜"]] +
                food_calories["水果"][午餐["水果"]] +
                food_calories["油脂"][午餐["油脂"]])

    晚餐熱量 = (food_calories["五穀雜糧"][晚餐["五穀雜糧"]] +
                food_calories["蛋白質"][晚餐["蛋白質"]] +
                food_calories["蔬菜"][晚餐["蔬菜"]] +
                food_calories["水果"][晚餐["水果"]] +
                food_calories["油脂"][晚餐["油脂"]])

    總每日熱量 = 早餐熱量 + 午餐熱量 + 晚餐熱量

  # return隨機產出的三餐餐點
    return 早餐, 午餐, 晚餐, 總每日熱量
```
