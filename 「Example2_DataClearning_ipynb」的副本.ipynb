```python
import pandas as pd
# read data and put it in a dataframe
# 在 google 工作表載入 gsheets
gsheets = gc.open_by_url('https://docs.google.com/spreadsheets/d/1V4c0lHTNIfSW3PX52pL4AkGCQ7x9YX3oqWf6pVlyXYA/edit?usp=sharing')
```

```python
# 從 gsheets 的 All-whiteboard-device 載入 sheets
sheets = gsheets.worksheet('anonymized_xapi').get_all_values()
# 將 sheets1 資料載入 pd 的 DataFrame 進行分析
df = pd.DataFrame(sheets[1:], columns=sheets[0])
# 取得最前面的5筆資料
df.head()
```

```python
# 設定alldays為時間戳最大-最小，也就是資料調查天數
df['timestamp'] = pd.to_datetime(df['timestamp'])
alldays = max(set(df['timestamp'].dt.date)) - min(set(df['timestamp'].dt.date))
```

```python
alldays.days
```

```python
len(df)
```

```python
# 將json檔展開成表格
import json
df_expanded = pd.json_normalize(df['json'].apply(json.loads))
```

```python
df_expanded
```

```python
# 資料格式
type(df_expanded['timestamp'][0])
```

```python
df_expanded['id'] = df['id']  # 加入每筆紀錄的 id（用來識別每一筆行為）
df_expanded['timestamp'] = pd.to_datetime(df['timestamp'])  # 將時間欄位轉換成「時間格式」，這樣才能進一步分析時間（例如：幾月幾號、幾點）
df_expanded['Date'] = df_expanded['timestamp'].dt.date   # 從時間中提取出「日期」部分，例如 2024-06-11
df_expanded['Time'] = df_expanded['timestamp'].dt.time    # 從時間中提取出「時間」部分，例如 22:17:50
df_expanded['Weekday'] = df_expanded['timestamp'].dt.day_name()  # 取得星期幾的名稱，例如 Monday、Tuesday，方便看行為是發生在哪一天
df_expanded.head()  # 顯示處理後的前五筆資料，幫助我們確認資料長得正不正確
```

```python
# 根據「學生名稱」和「日期」把資料分組（每個學生每天一組）
grouped = df_expanded.groupby(['actor.name', 'Date']).agg({

    # 把每一天的時間戳記都收集起來，變成一個列表
    'timestamp': list,

    # 把每一天學生做過的動作（例如：觀看、回答）都收集成一個列表
    'verb.display.zh-TW': list,

    # 把每一天互動過的學習資源（例如影片、題目）的 ID 收集起來
    'object.id': list

# 聚合完資料後，重設索引，把分組的欄位變成正常欄位
}).reset_index()
```

```python
# 把「日期」的資料格式轉換成程式能理解的時間格式（方便後面做處理）
grouped['Date'] = pd.to_datetime(grouped['Date'])

# 從日期中找出是星期幾，新增一欄叫做「Weekday」（例如：Monday、Tuesday）
grouped['Weekday'] = grouped['Date'].dt.day_name()
grouped
```

```python
len(set(grouped['actor.name'])) #有297位學生
```

根據每位學生的瀏覽日期，來去畫出不同學生瀏覽影片的天數

```python
# 計算每個學生（actor.name）有「幾天」有活動紀錄（用 nunique() 計算不重複的天數）
student_days = grouped.groupby('actor.name')['Date'].nunique()

# 嘗試幫結果命名欄位
student_days.colnames = ['actor.name', 'Date']

# 把索引轉回欄位，讓結果變成一個普通的表格
student_days = student_days.reset_index()
student_days
```

```python
student_days[student_days['Date'] > (alldays.days / 4)]['actor.name']
```

```python
import plotly.express as px
# 計算每個 Date 值出現多少人
date_counts = student_days['Date'].value_counts().reset_index()
date_counts.columns = ['Date', 'Count']

# 使用 Plotly 繪製條形圖
fig = px.bar(date_counts.sort_values(by='Date'),
             x='Date', y='Count',
             labels={'Date': '天數', 'Count': '人數'},
             title='每個天數對應的人數')

# 顯示圖表
fig.show()
```

```python
# 只看學生315的行為
grouped[grouped['actor.name'] == '學生315']['verb.display.zh-TW']
```

```python
# 只看「學生315」做過哪些動作，並統計每個動作出現幾次（像是觀看幾次、回答幾次）
grouped[grouped['actor.name'] == '學生315']['verb.display.zh-TW'].value_counts()
```

查看同學不同的行為，他們總共有哪些動作

```python
# 把每一天每位學生做過的動作去除重複，只保留「不重複」的動作（轉成 set 集合）
grouped['unique_verbs'] = grouped['verb.display.zh-TW'].apply(lambda x: set(x))

# 把所有學生每天的「不重複動作」打開攤平（explode），變成一個一個的動作列
unique_types = grouped['unique_verbs'].explode()

# 再次轉成 set（集合）來移除重複，得到「所有學生一共出現過哪些動作」
distinct_types = set(unique_types)

distinct_types
```

透過這個程式碼我們可以調閱不同學生他們不同動作的記數

```python
# 先挑出學生315的資料
student = grouped[grouped['actor.name'] == '學生315']

# 把他所有的動作串起來（是 list of list，要攤平成一個大 list）
all_verbs = student['verb.display.zh-TW'].explode()

# 計算每個動作的出現次數
verb_counts = all_verbs.value_counts()

# 顯示結果
verb_counts
```
