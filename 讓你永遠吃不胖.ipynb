```python
import pandas as pd
import random

# 根據政府提供資料填入不同食物的熱量讓電腦知道(當然也可以針對自己常吃的東西填入相應熱量)
food_calories = {

    "五穀雜糧": {
        "米": 200,
        "麵": 220,
        "麥片": 150,
        "地瓜": 90,
        "玉米": 100,
        "全麥麵包": 120
    },
    "蛋白質": {
        "蛋": 70,
        "豆腐": 80,
        "雞胸肉": 165,
        "魚": 140,
        "蝦": 90,
        "豬肉": 210,
        "豆類": 130
    },
    "乳製品": {
        "牛奶": 120,
        "起司": 90,
        "優格": 150
    },
    "蔬菜": {
        "菠菜": 25,
        "花椰菜": 55,
        "紅蘿蔔": 50,
        "羽衣甘藍": 40,
        "萵苣": 15,
        "小黃瓜": 20
    },
    "水果": {
        "香蕉": 105,
        "蘋果": 95,
        "橘子": 62,
        "西瓜": 85,
        "木瓜": 55,
        "奇異果": 42
    },
    "油脂": {
        "橄欖油": 120,
        "芝麻油": 120,
        "杏仁": 100,
        "花生": 90,
        "核桃": 100,
        "葵花籽": 80
    }
}
```

## 建個函式幫我們隨機創出三餐的menu，並記錄卡路里

```python

def generate_daily_meal():
    早餐 = {
        "五穀雜糧": random.choice(list(food_calories["五穀雜糧"].keys())),
        "蛋白質": random.choice(list(food_calories["蛋白質"].keys())),
        "乳製品": random.choice(list(food_calories["乳製品"].keys())),
        "水果": random.choice(list(food_calories["水果"].keys()))
    }

    午餐 = {
        "五穀雜糧": random.choice(list(food_calories["五穀雜糧"].keys())),
        "蛋白質": random.choice(list(food_calories["蛋白質"].keys())),
        "蔬菜": random.choice(list(food_calories["蔬菜"].keys())),
        "水果": random.choice(list(food_calories["水果"].keys())),
        "油脂": random.choice(list(food_calories["油脂"].keys()))
    }

    晚餐 = {
        "五穀雜糧": random.choice(list(food_calories["五穀雜糧"].keys())),
        "蛋白質": random.choice(list(food_calories["蛋白質"].keys())),
        "蔬菜": random.choice(list(food_calories["蔬菜"].keys())),
        "水果": random.choice(list(food_calories["水果"].keys())),
        "油脂": random.choice(list(food_calories["油脂"].keys()))
    }

# 計算每餐的總熱量
    早餐熱量 = (food_calories["五穀雜糧"][早餐["五穀雜糧"]] +
                food_calories["蛋白質"][早餐["蛋白質"]] +
                food_calories["乳製品"][早餐["乳製品"]] +
                food_calories["水果"][早餐["水果"]])

    午餐熱量 = (food_calories["五穀雜糧"][午餐["五穀雜糧"]] +
                food_calories["蛋白質"][午餐["蛋白質"]] +
                food_calories["蔬菜"][午餐["蔬菜"]] +
                food_calories["水果"][午餐["水果"]] +
                food_calories["油脂"][午餐["油脂"]])

    晚餐熱量 = (food_calories["五穀雜糧"][晚餐["五穀雜糧"]] +
                food_calories["蛋白質"][晚餐["蛋白質"]] +
                food_calories["蔬菜"][晚餐["蔬菜"]] +
                food_calories["水果"][晚餐["水果"]] +
                food_calories["油脂"][晚餐["油脂"]])

    總每日熱量 = 早餐熱量 + 午餐熱量 + 晚餐熱量

  # return隨機產出的三餐餐點
    return 早餐, 午餐, 晚餐, 總每日熱量
```

## 如何將產生的餐點menu匯入google表單

```python
# 生成一個月的餐點
days_in_month = 30
meals = []

for day in range(1, days_in_month + 1):
    早餐, 午餐, 晚餐, 每日總熱量 = generate_daily_meal()
    meals.append([day, "早餐", 早餐["五穀雜糧"], 早餐["蛋白質"], 早餐["乳製品"], 早餐["水果"], "", "", sum([food_calories["五穀雜糧"][早餐["五穀雜糧"]], food_calories["蛋白質"][早餐["蛋白質"]], food_calories["乳製品"][早餐["乳製品"]], food_calories["水果"][早餐["水果"]]])])
    meals.append([day, "午餐", 午餐["五穀雜糧"], 午餐["蛋白質"], "", 午餐["蔬菜"], 午餐["水果"], 午餐["油脂"], sum([food_calories["五穀雜糧"][午餐["五穀雜糧"]], food_calories["蛋白質"][午餐["蛋白質"]], food_calories["蔬菜"][午餐["蔬菜"]], food_calories["水果"][午餐["水果"]], food_calories["油脂"][午餐["油脂"]]])])
    meals.append([day, "晚餐", 晚餐["五穀雜糧"], 晚餐["蛋白質"], "", 晚餐["蔬菜"], 晚餐["水果"], 晚餐["油脂"], sum([food_calories["五穀雜糧"][晚餐["五穀雜糧"]], food_calories["蛋白質"][晚餐["蛋白質"]], food_calories["蔬菜"][晚餐["蔬菜"]], food_calories["水果"][晚餐["水果"]], food_calories["油脂"][晚餐["油脂"]]])])
    meals.append([day, "每日總熱量", "", "", "", "", "", "", 每日總熱量])

# 定義表頭
header = ["日期", "餐別", "五穀雜糧", "蛋白質", "乳製品", "蔬菜", "水果", "油脂", "熱量"]
```
